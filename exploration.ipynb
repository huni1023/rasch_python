```python
import os
import pickle

with open('data.pkl', 'rb') as f:
    data = pickle.load(f)
data[:10]
```

```python
len(data)
```

```python
import pandas as pd
d_dict = {'student_id': list(),
          'question_id': list(),
          'response': list()}

for (stu_id, que_id, res) in data:
    d_dict['student_id'].append(stu_id)
    d_dict['question_id'].append(que_id)
    d_dict['response'].append(res)

df = pd.DataFrame.from_dict(d_dict)
df.head(7)
```

```python
df.shape
```

```python
df.to_excel('data_daehanKim.xlsx', index=False)
```
